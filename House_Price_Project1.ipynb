### Determining the variables of interes

```python
# I will remove the 'Neighborhood', 'Exterior1st', 'Exterior2nd' variables, because their have a lot of unuiqe value
raw_data.drop(['Neighborhood', 'Exterior1st', 'Exterior2nd'], axis=1, inplace=True)
```

### Dealing with Missing Values

```python
pd.options.display.max_rows = 100
pd.options.display.max_columns = 100
```

```python
raw_data.isnull().sum()
```

```python
# Let's look at the percentage of missing values
(raw_data.isnull().sum() / len(raw_data)) * 100
```

```python
# Drop columns that have a lot of missing values (Alley, PoolQC, Fence, MiscFeature)
raw_data.drop(['Alley', 'PoolQC', 'Fence', 'MiscFeature'], axis=1, inplace=True)
```

```python
raw_data.shape
```

```python
## Fill missing values
for i in raw_data.columns:
    if raw_data[i].dtypes == 'int64' or raw_data[i].dtypes == 'float64':
        raw_data[i] = raw_data[i].fillna(raw_data[i].mean())
    if raw_data[i].dtypes == 'object':
        raw_data[i] = raw_data[i].fillna(raw_data[i].mode()[0])
```

```python
## Here I present the correlation between numrical variables and 'LogSalePrice'
pd.options.display.max_rows = 200
corr = data_cleaned.corr()
df_corr = pd.DataFrame(corr['LogSalePrice'])
df_corr.sort_values(by=['LogSalePrice'], ascending=False)
```

```python
## Here I show the variables that don't have correlation and I will drop from my original dataset
no_corr_var = df_corr[df_corr.values < 0.543190]
v = no_corr_var.index.values
```

```python
data_correlated = data_cleaned.drop(v, axis=1)
```

```python
data_correlated.head()
```

```python
## Final result of correlation between price and features
corr = data_correlated.corr()
df_corr = pd.DataFrame(corr['LogSalePrice'])
df_corr.sort_values(by=['LogSalePrice'], ascending=False)
```
